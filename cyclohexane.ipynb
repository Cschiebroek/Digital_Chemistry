```python
import pandas as pd
df = pd.read_csv('/home/cschiebroek/Downloads/13065_2015_134_MOESM1_ESM - Sheet1.csv')
df
smiles = {
    'water': 'O',
    'ethanol': 'CCO',
    '1-propanol': 'CCCO',
    'acetone': 'CC(C)=O',
    't-butanol': 'CC(C)(C)O',
    '2-methyl-1-propanol': 'CC(C)CO',
    '1-butanol': 'CCCCO',
    'neon' : 'Ne',
    'argon' : 'Ar',
    'nitrogen' : 'N#N',
    'krypton' : 'Kr',
    'methane' : 'C',
    'xenon' : 'Xe', 
    'sulphur hexafluoride' : 'FS(F)(F)(F)(F)F',
    'benzene' : 'c1ccccc1',
    'toluene' : 'Cc1ccccc1',
    'ethylbenzene' : 'CCc1ccccc1',
    'toluene' : 'Cc1ccccc1',
    'cyclohexane' : 'C1CCCCC1',
    'methylcyclohexane' : 'CC1CCCCC1',
    'styrene' : 'C=Cc1ccccc1',
}
df['SMILES'] = df['name'].map(smiles)
df
```

```python
#drop everything that is inorganic (no carbon)
inorganic_smiles = ['Ne', 'Ar', 'N#N', 'Kr', 'C', 'Xe', 'FS(F)(F)(F)(F)F']
df = df[~df['SMILES'].isin(inorganic_smiles)]
df
#also drop water as we can not make a graph if we exclude hydrgoens
df = df[df['SMILES'] != 'O']
```

```python
from rdkit.Chem import PandasTools
PandasTools.AddMoleculeColumnToFrame(df, smilesCol='SMILES', molCol='mol')
PandasTools.FrameToGridImage(df, column='mol', legendsCol="name", molsPerRow=4, subImgSize=(400, 200))
```

```python
#now subdf with only smiles and logP_mcy
subdf = df[['SMILES', 'logP_mcy']]
subdf
#to csv
subdf.to_csv('logP_mcy.csv', index=False)
```

```python
#now do leave-one-out; train the model on all but one of the smiles and predict the logP_mcy for the left out one
import os
preds,trues = [],[]
for smi in subdf['SMILES'].values:
    df_train = df_combined[df_combined['SMILES'] != smi]
    df_test = df_combined[df_combined['SMILES'] == smi]
    df_train.to_csv('train_leave_one_out.csv', index=False)
    df_test.to_csv('test_leave_one_out.csv', index=False)
    # train MTL
    os.system(f'''chemprop train \
    --data-path train_leave_one_out.csv \
    --task-type regression \
    --epochs 50 \
    --output-dir "cyclohexane_leave_one_out/{smi}" \
    --num-workers 15 ''')

    # Predict
    os.system(f'''chemprop predict \
    --test-path test_leave_one_out.csv \
    --model-path cyclohexane_leave_one_out/{smi}/model_0/best.pt \
    --preds-path "cyclohexane_leave_one_out/{smi}/preds.csv" \
    --num-workers 15''')
```

```python
#load all the predictions
import pandas as pd
from modules.utils_data import load_data
import numpy as np
import matplotlib.pyplot as plt

subdf = pd.read_csv('logP_mcy.csv')
df_combined = pd.read_csv('df_all_incl_cyclohexane.csv')
preds,trues = [],[]

for smi in subdf['SMILES'].values:
    try:
        df_train = pd.read_csv(f'/localhome/cschiebroek/PhysPropMTL/Digital_Chemistry/cyclohexane_leave_one_out/{smi}/model_0/test_predictions.csv')
        hex_idx = df_train.columns.get_loc('logP_mcy') -1
        df_pred = pd.read_csv(f'/localhome/cschiebroek/PhysPropMTL/Digital_Chemistry/cyclohexane_leave_one_out/{smi}/preds_0.csv')
        val = df_pred['logP_mcy'].values[0]
        pred = df_pred[f'pred_{hex_idx}'].values[0]
        preds.append(pred)
        trues.append(val)
        # print(f'Predicted: {pred}, True: {val})')
        if abs(pred-val) > 1:
            print(f'Predicted: {pred}, True: {val}, SMILES: {smi}')
    except FileNotFoundError:
        continue
#get stats, r2, mae, rmse, kendall tau
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
import scipy.stats
r2 = r2_score(trues, preds)
mae = mean_absolute_error(trues, preds)
rmse = np.sqrt(mean_squared_error(trues, preds))
tau, p_value = scipy.stats.kendalltau(trues, preds)
plt.scatter(trues, preds)
plt.xlabel('True logP_mcy')
plt.ylabel('Predicted logP_mcy')
plt.title('Leave-one-out predictions for cyclohexane')
textstr = f'RMSE: {rmse:.2f}\nMAE: {mae:.2f}\n$R^2$: {r2:.2f}\nKendall tau: {tau:.2f}'
props = dict(boxstyle='round', facecolor='wheat', alpha=0.5)
plt.text(0.05, 0.95, textstr, transform=plt.gca().transAxes, fontsize=10, verticalalignment='top', bbox=props)
#draw diagonal
plt.plot([min(trues), max(trues)], [min(trues), max(trues)], 'k--')
#and +/- 0.5 logP units
plt.plot([min(trues), max(trues)], [min(trues)+1, max(trues)+1], 'r--')
plt.plot([min(trues), max(trues)], [min(trues)-1, max(trues)-1], 'r--')

plt.show()
```

```python
#load all the predictions
import pandas as pd
from modules.utils_data import load_data
import numpy as np
import matplotlib.pyplot as plt

preds,trues = [],[]

for fold in range(5):
        try:
            df_pred = pd.read_csv(f'cyclohexane_leave_one_out/fold_{fold}/preds_0.csv')
            preds_logp = df_pred['LogP_cyclohexane'].values
            smiles = df_pred['SMILES'].values
            for i in range(len(smiles)):
                df_tmp = df_combined[df_combined['SMILES'] == smiles[i]]
                true_logp = df_tmp['LogP_cyclohexane'].values[0]
                preds.append(preds_logp[i])
                trues.append(true_logp)      
                print(f'Predicted: {preds_logp[i]}, True: {true_logp})')
        except FileNotFoundError:
            continue

    
#get stats, r2, mae, rmse, kendall tau
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
import scipy.stats
r2 = r2_score(trues, preds)
mae = mean_absolute_error(trues, preds)
rmse = np.sqrt(mean_squared_error(trues, preds))
tau, p_value = scipy.stats.kendalltau(trues, preds)
plt.scatter(trues, preds)
plt.xlabel('True logP_mcy')
plt.ylabel('Predicted logP_mcy')
plt.title('KFold predictions for cyclohexane')
textstr = f'RMSE: {rmse:.2f}\nMAE: {mae:.2f}\n$R^2$: {r2:.2f}\nKendall tau: {tau:.2f}'
props = dict(boxstyle='round', facecolor='wheat', alpha=0.5)
plt.text(0.75, 0.2, textstr, transform=plt.gca().transAxes, fontsize=10, verticalalignment='top', bbox=props)
#draw diagonal
# plt.plot([min(trues), max(trues)], [min(trues), max(trues)], 'k--')
#and +/- 0.5 logP units
# plt.plot([min(trues), max(trues)], [min(trues)+1.5, max(trues)+1.5], 'r--')
# plt.plot([min(trues), max(trues)], [min(trues)-1.5, max(trues)-1.5], 'r--')
#make square
plt.gca().set_aspect('equal', adjustable='box')
#and +/- 1 logP units
plt.xlim(min(trues)-1, max(trues)+1)
plt.ylim(min(trues)-1, max(trues)+1)
plt.plot([min(trues)-1, max(trues)+1], [min(trues)-1, max(trues)+1], 'k--')
plt.plot([min(trues)-1, max(trues)+1], [min(trues)-2.5, max(trues)-0.5], 'r--')
plt.plot([min(trues)-1, max(trues)+1], [min(trues)+0.5, max(trues)+2.5], 'r--')
#and +- 1.5 logP units
#also get pearson r
from scipy.stats import pearsonr
r, p_value = pearsonr(trues, preds)
print(f'Pearson r: {r:.2f}')

plt.show()
```
