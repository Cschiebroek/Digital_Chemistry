```python
import pandas as pd
from sklearn.model_selection import KFold
import os
train_df = pd.read_csv('train_df_cyclohexane.csv')
test_df = pd.read_csv('test_df_cyclohexane.csv')
kf = KFold(n_splits=5, shuffle=True, random_state=42)
kf.get_n_splits(test_df)
print(kf)
props = train_df.columns[1:].tolist()
for i, (train_index, test_index) in enumerate(kf.split(test_df)):
    print(f'Fold {i}')
    train, test = test_df.iloc[train_index], test_df.iloc[test_index]
    #combine train with train_df
    train = pd.concat([train, train_df])
    train.to_csv(f'train_leave_one_out_fold_{i}.csv', index=False)
    test.to_csv(f'test_leave_one_out_fold_{i}.csv', index=False)


    os.system(f'''chemprop train \
    --data-path train_leave_one_out_fold_{i}.csv \
    --task-type regression \
    --epochs 50 \
    --output-dir "cyclohexane_leave_one_out/fold_{i}" \
    --target-columns {" ".join(props)} \
    --num-workers 15 ''')

    # Predict
    os.system(f'''chemprop predict \
    --test-path test_leave_one_out_fold_{i}.csv \
    --model-path cyclohexane_leave_one_out/fold_{i}/model_0/best.pt \
    --preds-path "cyclohexane_leave_one_out/fold_{i}/preds.csv" \
    --target-columns {" ".join(props)} \
    --num-workers 15''')
```
